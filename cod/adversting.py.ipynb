```python
#splitting the data for training and testing
from sklearn.model_selection import train_test_split
A_train, A_test, a_train, b_test = train_test_split(sales_df[['TV']], sales_df[['Sales']], test_size = 0.3,random_state=0)
```

```python
#applying linear regression for best regression line
from sklearn.linear_model import LinearRegression
type1 = LinearRegression()
type1.fit(A_train,a_train,)
```

```python
#Prediction
result_Prediction= type1.predict(A_test)
print(result_Prediction)
```
